```python
import cv2
import torch
import numpy as np
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', trust_repo=True)

from models.experimental import attempt_load
def scale_coords(img_shape, coords, img0_shape, ratio_pad=None):
    # Rescale coords (xyxy) from img1_shape to img0_shape
    # ratio_pad = (w_ratio, h_ratio, w_padding, h_padding)
    pad_w = pad_h = 0
    if ratio_pad:
        # Ratio padding https://github.com/ultralytics/yolov5/issues/1487
        r_w, r_h, pad_w, pad_h = ratio_pad
        coords[:, [0, 2]] -= pad_w  # x padding
        coords[:, [1, 3]] -= pad_h  # y padding
        coords[:, [0, 2]] /= r_w  # width ratio
        coords[:, [1, 3]] /= r_h  # height ratio
    gain = min(img_shape[0] / img0_shape[0], img_shape[1] / img0_shape[1])  # gain  = old / new
    pad = (img_shape[1] - img_shape[0]) / 2 if img_shape[0] < img_shape[1] else (
            img_shape[0] - img_shape[1]) / 2
    coords[:, [0, 2]] -= pad  # x padding
    coords[:, [1, 3]] -= pad  # y padding
    coords[:, :4] /= gain
    clip_coords(coords, img0_shape)
    coords[:, [0, 2]] += pad_w  # x padding
    coords[:, [1, 3]] += pad_h  # y padding
    return coords.round().astype(int)
from utils.general import non_max_suppression
# from utils.plots import plot_one_box, colors

# import letterbox
from models.common import letterbox

#
# from utils.general import non_max_suppression, scale_coords
# from utils.plots import plot_one_box
```
